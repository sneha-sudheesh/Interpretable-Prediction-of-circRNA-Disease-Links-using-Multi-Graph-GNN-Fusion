```python
import torch
from pathlib import Path

from sage_encoder import GraphSAGEEncoder


ROOT = Path().resolve().parents[1]

GRAPH_DIR = ROOT / "graphs"
EMB_DIR = ROOT / "embeddings"
EMB_DIR.mkdir(exist_ok=True)


# --------------------------
# Load graphs
# --------------------------

gcm = torch.load(GRAPH_DIR / "gcm_graph.pt", weights_only=False)
gmd = torch.load(GRAPH_DIR / "gmd_graph.pt", weights_only=False)
gcd = torch.load(GRAPH_DIR / "gcd_graph_fold0.pt", weights_only=False)


# --------------------------
# Encoder
# --------------------------

encoder = GraphSAGEEncoder(
    in_channels=6,
    hidden_channels=64,
    out_channels=64
)

encoder.eval()


# --------------------------
# Forward pass
# --------------------------

with torch.no_grad():

    Ecm = encoder(gcm.x, gcm.edge_index)
    Emd = encoder(gmd.x, gmd.edge_index)
    Ecd = encoder(gcd.x, gcd.edge_index)


# --------------------------
# Save embeddings
# --------------------------

torch.save(Ecm, EMB_DIR / "Ecm_fold0.pt")
torch.save(Emd, EMB_DIR / "Emd.pt")
torch.save(Ecd, EMB_DIR / "Ecd_fold0.pt")


# --------------------------
# Print shapes
# --------------------------

print("Ecm:", Ecm.shape)
print("Emd:", Emd.shape)
print("Ecd:", Ecd.shape)
```
